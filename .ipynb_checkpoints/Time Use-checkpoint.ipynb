<h3> Is there a significant difference between the way that working men and women spend their day on average? This could potentially impact their wages and how much they put childcare before work. <h3/>

<h4>In order to see real statistical difference we run two sample t test on difference in means. 
Null hypothesis: 2 independent samples have identical average (expected) values. We reject the null if our p-value from the test is less than our alpha = .05 <h4/>

```python
#statistical analysis requires standard error, not just means... why I'm adding more detailed data from BLS

import os
import pandas as pd
import numpy as np
import glob
import sys
from scipy import stats
#print(sys.version)
```

```python
df = pd.DataFrame()
doc_root = os.path.join(os.environ['HOME'],'Downloads/BLS_Data')
files_xls = glob.glob(os.path.join(doc_root, '*.xlsx'))
files_xls
for f in files_xls:
    #print f
    variable = os.path.splitext(os.path.basename(f))[0].split('-')[1]
    #print variable
    frame = pd.read_excel(f, header = 13)
    frame = frame.drop('Unnamed: 4', 1)
    frame = frame.drop('Unnamed: 5', 1)
    #print frame
    frame['Variable'] = variable
    df = df.append(frame)

#df
```

<h4>Employed Men Vs. Employed Women - hours spent working : <h4/>

```python
women_working = df[df['Variable'] == 'Working, Employed, Women']
men_working = df[df['Variable'] == 'Working, Employed, Men']
women_working_values = women_working['Estimate']
men_working_values = men_working['Estimate']
alpha = .05
t_stat, p_val = stats.ttest_ind(women_working_values, men_working_values)
p_val
```

<h4>Employed Men Vs. Employed Women - hours spent on household activities : <h4/>

```python
#all of these are for employed -- need to load more data for unemployed
women_working = df[df['Variable'] == 'Household activities (includes travel), Women']
men_working = df[df['Variable'] == 'Household activities (includes travel), Men']
women_working_values = women_working['Estimate']
men_working_values = men_working['Estimate']
t_stat, p_val = stats.ttest_ind(women_working_values, men_working_values)
p_val
```

<h4>Employed Men Vs. Employed Women - hours spent caring for household children : <h4/>

```python
women_working = df[df['Variable'] == 'Caring for and helping household children, Parents of household children, Women']
men_working = df[df['Variable'] == 'Caring for and helping household children, Parents of household children, Men']
women_working_values = women_working['Estimate']
men_working_values = men_working['Estimate']
t_stat, p_val = stats.ttest_ind(women_working_values, men_working_values)
p_val
```

<h5> In all the above cases, our p values were very small, and therefore we are able to reject the null hypothesis. There is a difference between the average number of hours spent on these activities for employed men and for employed women. <h5/>
