```python
def query_url(term):
    initial_url = 'http://www.urbandictionary.com/'
    #term = '/define.php?term=' + term + "?"
    full_url = initial_url + term 
    page = requests.get(full_url)
    page_html = page.text
    soup = BeautifulSoup(page_html, 'lxml')
    text = soup.find_all(name='div',attrs={'class':'meaning'})
    return text
```
